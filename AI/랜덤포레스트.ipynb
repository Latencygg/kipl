# Random forest with Grams

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

def auto_random_forest(train_x, train_y, test_x, test_y, max_depth_range=range(10, 50)):
    result = []
    # max_depth_range 최대 트리 깊이의 범위
    
    # criterion, max_feature, min_amples,split, max_depth 하이퍼파라미터 설정
    for criterion in ["gini", "entropy", "log_loss"]:
    # 분할 기준으로 지니 불순도, 엔트로피, 로그손실 중 설정
    
        for max_feature in ["sqrt", "log2", None]:
        # 각 노트에서 고려할 특성으로 sqrt, log2, None 중 설정
        # sqrt : 특성 개수의 제곱근만큼 특성을 무작위로 선택
        # log2 : 특성 개수의 로그 값을 밑으로 2를 적용한 값만큼 특성을 무작위로 선택
        # 예를 들어, 16개의 특성이 있으면 log2의 16인 4개의 특성을 무작위로 선택
        # None : 특성을 무작위로 선택하지 않음
        
            for min_samples_split in range(2, 10):
            # 노드를 분할하기 위해 필요한 최소 샘플 수
            # 작은 값으로 설정하면 더 많은 노드가 생성되고 모델이 더 복잡해짐
            # 모델이 복잡해질수록 새로운 데이터에 대한 일반화 능력이 떨어짐
            
                for max_depth in max_depth_range:
                # 트리의 최대 깊이
                # 트리가 깊어질수록 모델이 학습 데이터에 과적합될 가능성이 높아짐
                    
                    # 랜덤포레스트 모델
                    model = RandomForestClassifier(
                        criterion=criterion,
                        max_depth=max_depth,
                        max_features=max_feature,
                        min_samples_split=min_samples_split,
                        n_jobs = -1
                        # n_jobs cpu 코어 수, -1로 설정하여 가능한 모든 코어 사용
                    )
                    model.fit(train_x, train_y)
                    
                    # 예측
                    train_pred = model.predict(train_x)
                    test_pred = model.predict(test_x)
                    
                    result.append({
                        "criterion" : criterion,
                        "max_depth" : max_depth,
                        "max_features" : max_feature,
                        "min_samples_splits" : min_samples_split,
                        # 정확도
                        "train_accuracy" : accuracy_score(train_pred, train_y),
                        "test_accuracy" : accuracy_score(test_pred, test_y)
                        })
    # result에서 테스트 정확도가 가장 높은 모델 탐색
    max_acc_idx = 0
    # 가장 높은 테스트 정확도를 가진 모델의 인덱스를 추적하기 위한 변수
    for idx, r in enumerate(result):
    # result의 인덱스와 결과 데이터를 가져옴
        if r["test_accuracy"] > result[max_acc_idx]["test_accuracy"]:
        # 현재 결과의 테스트 정확도와 현재까지 가장 높은 테스트 정확도 비교
        # 현재 결과의 테스트 정확도가 더 높다면 max_acc_idx값을 현재 인덱스 idx로 업데이트
        # 가장 높은 테스트 정확도를 가진 결과의 인덱스를 업데이트
            max_acc_idx = idx
    
    return result, max_acc_idx
```

```python
result, max_idx = auto_random_forest(x_train, y_train, x_val, y_val)
```
